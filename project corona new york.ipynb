```python
total_tweet = []
cnt_sentiment_n = []
date=["2020-04-22","2020-04-23","2020-04-24","2020-04-25","2020-04-26","2020-04-27"]
query = ['covid-19 -filter:retweets']
max_tweetsn = 1000

count1n = 0
for d in date:
    try :


        print("Fetching data from api For New York")
        tweetsn = tweepy.Cursor(api.search,q=query,language='en',geocode = "40.7128,74.060,500mi",until=d).items(max_tweets)
        print("Data fetched")

        print("Begning to parse data for sentiment analysis")
        for tweet in tweetsn:   
            
                print(tweet.created_at)
   
                total_tweet.append({"text":tweet.text,"created_at":tweet.created_at.strftime("%x")})
                count1n=count1n+1
    
                if(count1n % 10 == 0):
                    print(str((count1n/max_tweetsn)*100)+"% done" )
   
                analysisn = TextBlob(tweet.text)
                if analysisn.sentiment[0]>0:
                     cnt_sentiment_n.append({"text":tweet.text,"Sentiment":"positive","created_at":tweet.created_at.strftime("%m-%d")})
        
                elif analysisn.sentiment[0]< 0:
                     cnt_sentiment_n.append({"text":tweet.text,"Sentiment":"negative","created_at":tweet.created_at.strftime("%m-%d")})

    except Exception as e: # work on python 2.x
    
        print("Error in fetching data"+str(e))

    finally:
        print("Completing process")
```
