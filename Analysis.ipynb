```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("Analysis of positive and negative tweets in California and New York. If california has more positive tweets or Ney York and \nvice versa. Also checking if positive tweets are more or negative tweets.")
```

```python
data_pdc = pd.read_csv("california_tweet_data3.csv") 
data_pdc.head()
```

```python
data_pdn = pd.read_csv("ny_tweet_data3.csv") 
data_pdn.head()
```

```python
df1, df2 = [a for _, a in data_pdc.groupby(data_pdc['Sentiment'] != 'negative')]
df1.reset_index()
df1.head()
```

```python
df2.reset_index()
df2.head()
```

```python
df3, df4 = [a for _, a in data_pdn.groupby(data_pdn['Sentiment'] != 'negative')]
df3.reset_index()
df3.head()
```

```python
df4.reset_index()
df4.head()
```

```python
y1=df1['Count'].tolist()
y2=df2['Count'].tolist()
y3=df3['Count'].tolist()
y4=df4['Count'].tolist()
x = df1['created_at'].tolist()

# multiple line plot
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.title("Comparison of positive and negative tweets of California and new York")
plt.ylabel("Count")
plt.xlabel("Dates")
plt.savefig("Flow_basedondates_Positive_Negative_ny.png")
```

```python
print("Analysis: \n\n")
print("Blue:CAlifornia Negative \nOrange: California positive \nGreen: New York Negative \nRed: New York Positive")
print("Analysis Complete")
print("For both California and New York positive tweets are more than negative.\n")
print("Analysed that New York has more positive tweets then California, only on 23rd there is somewhat similar data for both. \nThis can be due to more positive cases reported in New York then in California. That's why people in New York are more \nconcerned.\n")
print("Negative tweets of New York and California are mixed. For first few days negative tweetsin California are more than New York, \nthen after 23rd noon negative tweets in New York are more")
```
